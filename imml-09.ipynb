**<h4><u>Write a Pandas program to create a plot of Open, High, Low, Close,Adjusted Closing prices and Volume of given company between two specific dates.</u></h4>**

```python
# Name: Salet Shubham B.
# Enrollment: 216270307030

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/kaggle/input/tesla-data2/NFLX.csv')

# det Dtae as dataframe column 
df.set_index('Date', inplace=True)

start_date = '2022-01-19'
end_date = '2022-02-04'

# specific range of date start to end
df = df.loc[start_date:end_date]

up = df[df.Close >= df.Open]
down = df[df.Close < df.Open]

# width for bars
width = .5
width2 = .1
  
plt.bar(up.index, up.Close-up.Open, width, bottom = up.Open)
plt.bar(up.index, up.High-up.Close, width2, bottom = up.Close)
plt.bar(up.index, up.Low-up.Open, width2, bottom = up.Open)
  
plt.bar(down.index, down.Close-down.Open, width, bottom=down.Open)
plt.bar(down.index, down.High-down.Open, width2, bottom=down.Open)
plt.bar(down.index, down.Low-down.Close, width2, bottom=down.Close)

plt.xticks(rotation='vertical')
plt.title('Stock Prices')
plt.xlabel('Date')
plt.ylabel('Price/Volume')
plt.grid(True)
plt.legend()
plt.show()
```
