**<h4><u>Write a Pandas program to create a line plot of the opening, closing stock prices of given company between two specific dates.</u></h4>**

```python
# Name: Salet Shubham B.
# Enrollment: 216270307030

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/kaggle/input/tesla-data/TSLA.csv')

start = "2010-06-29"
close = "2010-07-13"
df = df[(df['Date'] >= start) & (df['Date'] <= close)]

df = df.set_index('Date').sort_index(ascending=True)
df = df.reset_index()

df_open = df[['Date', 'Open']]
df_close = df[['Date', 'Close']]

plt.figure(figsize=(10, 5))
plt.plot(df_open['Date'], df_open['Open'], color='green', marker='o', label='Open')

plt.plot(df_close['Date'], df_close['Close'], color='red', marker='o', label='Close')

plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Open and Close Prices')
plt.legend()
plt.xticks(rotation=60)
plt.grid(True)
plt.show()
```
